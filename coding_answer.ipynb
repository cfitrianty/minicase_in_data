```python
# Menambahkan format waktu pada trx_date
from faker import Faker
import datetime

fake = Faker()

# mendapatkan tanggal saat ini
current_date = datetime.datetime.now()

# Hitung tiga tahun sebelumnya
three_years_ago = current_date - datetime.timedelta(days=365 * 3)  # Mengurangkan 3 tahun dalam satuan hari


```

```python
# Rename library faker() & buat data table
fake = Faker()
data = []

for _ in range(1000000):
    account_number = f'10{fake.random_number(digits=4)}'
    trx_type = random.choice(information_option)
    amount = round(random.uniform(100, 100000))
    flag_db_cr = random.choice(['C', 'D'])
    information = f'{trx_type} {fake.random_int(min=1000000, max=9999999)}'
    trx_date = fake.date_between(start_date= three_years_ago, end_date='today')
    reference_number = fake.random_int(min=100000, max=999999)
    currency = 'IDR'
    trx_time = fake.time()
    cif = fake.random_int(min=1000000000, max=9999999999)
    fitur = trx_categories.get(trx_type, "OTHER")
    city = fake.random_element(['Bandung', 'Jakarta', 'Solo', 'Samarinda', 'Riau'])
    gender = fake.random_element(['Male', 'Female'])
    age = fake.random_int(min=18, max=80)
    active_member = random.choice([0, 1])
    has_creditcard = random.choice([0, 1])
    exited = random.choice([0, 1])

    data.append({
        "ACCOUNT_NUMBER": account_number,
        "TRX_TYPE": trx_type,
        "AMOUNT": amount,
        "FLAG_DB_CR": flag_db_cr,
        "INFORMATION": information,
        "TRX_DATE": trx_date.strftime('%Y-%m-%d'),
        "REFERENCE_NUMBER": reference_number,
        "CURRENCY": currency,
        "TRX_TIME": trx_time,
        "CIF": cif,
        "FITUR": fitur,
        "CITY" : city,
        "GENDER" : gender,
        "AGE" : age,
        "ACTIVE_MEMBER" : active_member,
        "HAS_CREDITCARD" : has_creditcard,
        "EXITED" : exited,
        })

# simpan dummy data ke csv file
csv_filename = 'dummy_transaction.csv'
with open(csv_filename, 'w', newline='') as csvfile:
    fieldnames = data[0].keys()
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # membuat header
    writer.writeheader()

    # looping menulis baris
    for record in data:
        writer.writerow(record)

# Save data dummy file ke Json
# with open('dummy_data_transaction.json', 'w') as json_file:
#    json.dump(data, json_file, indent=4)

print("Data Dummy berhasil disimpan dalam 'dummy_transaction.csv'")
```

### Data Preprocessing

```python
# city (Frequency Encoding)
freq = data['city'].value_counts()
data['city'] = data['city'].map(freq)

# trx_type (Frequency Encoding)
freq = data['trx_type'].value_counts()
data['trx_type'] = data['trx_type'].map(freq)

# fitur (Frequency Encoding)
freq = data['fitur'].value_counts()
data['fitur'] = data['fitur'].map(freq)

# gender (Frequency Encoding)
freq = data['gender'].value_counts()
data['gender'] = data['gender'].map(freq)

# flag_db_cr (Binary Encoding)
data['flag_db_cr'] = data['flag_db_cr'].map({'d': 0, 'c': 1})

## trx_date (ubah tipe menjadi waktu)
# data['trx_date'] = pd.to_datetime(data['trx_date'])

## pisahkan berdasar tahun, bulan, tanggal
# data['x_trx'] = data['trx_date'].dt.year
# data['m_trx'] = data['trx_date'].dt.month
# data['d_trx'] = data['trx_date'].dt.day
## drop kolom trx_date
# data.drop(['trx_date'], axis=1)


data.info()
```

## Split train & test data

```python
from sklearn.model_selection import train_test_split

x = data.drop(['exited'], axis=1)
y = data['exited']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Scaling di train data setelah split data
from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()
x_train_std = scale.fit_transform(x_train)
x_train = pd.DataFrame(data=x_train_std, columns=x_train.columns)
x_train
```

```python
# Random Forest with Balanced Data

from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=42)
rfc.fit(x_train, y_train)
pre_rfc = rfc.predict(x_test)

print('Akurasi',accuracy_score(y_test, pre_rfc))
```

```python
# KNN model
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
pre_knn = knn.predict(x_test)

print('Akurasi',accuracy_score(y_test, pre_knn))
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=42)
lr_model = lr.fit(x_train, y_train)
pre_lr = lr.predict(x_test)
accuracy = accuracy_score(y_test, pre_lr)

print('Akurasi',accuracy_score(y_test, pre_lr))
```

```python
conditions = [
    {'trx_type': 'withdrawl transfer (atm)', 'fitur': 'payment'},
    {'trx_type': 'gopay', 'fitur': 'e-wallet'},
    {'trx_type': 'shopee', 'fitur': 'e-wallet'},
    {'trx_type': 'biaya admin', 'fitur': 'admin fee'},
    {'trx_type': 'payment telco pascabayar', 'fitur': 'payment'},
    {'trx_type': 'deposit cross border', 'fitur': 'cross-border deposit'},
    {'trx_type': 'withdrawl transfer', 'fitur': 'transfer'},
    {'trx_type': 'qris', 'fitur': 'e-wallet'},
    {'trx_type': 'payment cc', 'fitur': 'credit card payment'},
    {'trx_type': 'reversal', 'fitur': 'reversal'},
    {'trx_type': 'edc', 'fitur': 'payment'},
    {'trx_type': 'deposit transfer (atm)', 'fitur': 'atm deposit'}
    ]

def determine_category(row):
    for condition in conditions:
        if row['trx_type'] == condition['trx_type'] and row['fitur'] == condition['fitur']:
            if row['trx_type'] == 'withdrawl transfer (atm)' or row['trx_type'] == 'payment telco pascabayar' 
                or row['trx_type'] == 'biaya admin' or row['trx_type'] == 'deposit cross border' and row['fitur'] == 'payment' :
                return 'payment'
            elif row['trx_type'] == 'gopay' or row['trx_type'] == 'shopee' and row['fitur'] == 'e-wallet':
                return 'instant payment'
            elif row['trx_type'] == 'payment cc' and row['fitur'] == 'credit card payment':
                return 'bill'
            elif row['trx_type'] == 'qris' and row['fitur'] == 'e-wallet':
                return 'shopping'
            elif row['trx_type'] == 'edc' and row['fitur'] == 'payment':
                return 'shopping'
            elif row['trx_type'] == 'deposit transfer (atm)' and row['fitur'] == 'atm deposit':
                return 'savings'
            elif row['trx_type'] == 'reversal' and row['fitur'] == 'reversal':
                return 'reversal'
    return 'other'

dataset['category_trx'] = dataset.apply(determine_category, axis=1)
```
